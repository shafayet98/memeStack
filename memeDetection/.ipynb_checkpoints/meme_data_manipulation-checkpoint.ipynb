```python
import json
import urllib.request
```

```python
def getData(filename):
    with open(filename,'r') as fp:
        return json.load(fp)
```

```python
jsonObj = getData('meme.json')
```

```python
total_memes_length = len(jsonObj.get("_default"))

print(total_memes_length)
print(jsonObj.get("_default").get("1").get("media"))
```

```python
media_url = []
for i in range(1,total_memes_length):
    media_url.append(jsonObj.get("_default").get(str(i)).get("media"))
```
